# 4. Principal Component Analysis

## a. dimensionality reduction

common issue: too many independent variables
- disadvantage is that info needs to be refined; cannot keep all that info - too many columns and too few rows (classic overfitting)

the more independent, the beter the ml algo performance - independence avoids overfitting

### feature elimination and feature extraction

feature elimiation - remove columns

feature extraction - combine columns using some function into groups of variables

Principal Component Analysis is most common feature extraction technique

- only uses linear equations for z; linear combinations of x variables

## b. Principal Component Analysis

eigan vectors - natural axis of data

eigen values - magnitude (length) of axis



## c. Co-variance Matrix

take x1 and x2, and scale it and find covariance matrix of scaled data

find covariance matrix of scaled data - simply finding the correlation matrix

once covariance matrix found, we've found eigen vectors


## d. Eigen Vectors and Values

eigen = unique

eigen vectors and values - natural axis and magnitude

all these steps are essentially putting an ellipsoid around the cloud of datapoints

finding eigen vectors and values is finding the axis and magnitude of ellipsoid

eigan vectors of covariance matrix = principle components of data

can use eigen vectors to create new variables, for example, z-1 and z-2
- z-1 is linear combination of x-1
- z-2 is linear combination of x-2

eigen vectors dictate what a-11 etc are:



essentially: take data, figure out eigen vectors aka axis, and allow the axis to define data

## e. Overall steps - PCA

1. scale
    - scaling and find covarriance matrix of independent varaiables = finding correlation matrix
    - only independed variables as we are trying to reduce dimensionality of variables
1. break into magnitude and direction 
    - putting together ellipsoid on data
    - find eigen vectors and values
    - eigen values - say how long ellipsis axis are; will be percentage of variation explained
