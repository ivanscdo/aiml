```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from PIL import Image
```

```python
data = pd.read_csv('../files/Melbourne_housing_FULL.csv')
data.head()
```

## Uni-variate Analysis

```python
data = data['Distance']
```

```python
len(data)
```

```python
# missing values
data.isnull() # boolean
data.isnull().sum()
```

```python
data = data.dropna() # drops all missing values
# data.isnull().sum()
```

```python
# uni-variate analysis - working with one variable
# easiest way to look at distrubutions is to plot them, through a histogram for ex
# must create bins for histogram
plt.hist(data, bins=50)
# plt.hist(data)
```

```python
# sns.distplot(data)
sns.distplot(data, hist=False) # displays only the line
```

```python
sns.violinplot(data) # same as distribution plot, but also shows reflection across x-axis
```

between class limits 4.81 and 9.62 (the peak), we have highest number of observations

next interval has similar number of observations

modal class close to mean and median

most of data < 20, concentrated to the left

allows a sense of how far typical person lives, outliers

```python
plt.figure(figsize=(20,10)) # 10 bins
plt.hist(data, color='g')
plt.axvline(data.mean(), color='m', linewidth=1)
plt.axvline(data.median(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(data.mode()[0], color='w', linestyle='dashed', linewidth=1)
```

```python
sns.distplot(data, hist_kws=dict(cumulative=True), kde_kws=dict(cumulative=True))
# cumulative distribution - see how rapidly the distances add up
# slope represents histrogram
# adds up areas under histograms
# the sharper the rise, the more the data points
```
