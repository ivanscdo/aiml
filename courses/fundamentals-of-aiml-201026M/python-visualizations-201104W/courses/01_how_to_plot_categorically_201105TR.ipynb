```python
import pandas as pd
import seaborn as sns # sam n seaborn from west wing
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline 
# print seaborn graphs in jupyter notebook

auto = pd.read_csv('Automobile.csv')
print(auto.head())
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True);
# example of categorical variable: vehicle's fuel type
```

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower']);
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower']);
# lower whisker = Q3 + 1.5 * IQ (inter-quartile range, Q3-Q1), upper whisker Q1 - 1.5IQR
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type']);
```

```python
sns.barplot(auto['body_style'], auto['horsepower'], hue=auto['fuel_type']);
```

```python
sns.countplot(auto['body_style']);
# how many cars in these categories
```

```python
sns.countplot(auto['body_style'], hue=auto['fuel_type']);
```

```python
sns.pointplot(auto['body_style'], auto['horsepower'], hue=auto['number_of_doors']);
```

```python
sns.catplot(x='fuel_type', y='horsepower', hue='number_of_doors', col='drive_wheels', data=auto, kind='box');
```

```python
sns.catplot(x='fuel_type', y='horsepower', hue='number_of_doors', col='drive_wheels', data=auto, kind='violin');
```

```python
sns.catplot(x='fuel_type', hue='number_of_doors', col='drive_wheels', data=auto, kind='count');
```

```python
sns.lmplot(y='horsepower', x='engine_size', data=auto);
# lm - linear model
# solid blue line is line of best fit for relationship that horsepower = constant + constant * insize
# blue shaded - 95% confidence interval
```

```python
sns.lmplot(y='horsepower', x='engine_size', hue='fuel_type', data=auto);
```
