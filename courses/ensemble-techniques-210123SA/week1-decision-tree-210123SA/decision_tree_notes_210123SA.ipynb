# Desision Tree - CART

## CART-based Decision Tree ~0:39

CART - Classification and Regression Tree

### Classification vs Regression

- In classification, target variable is a categorical variable, yes/no buckets.
    - Each leaf will correspond to a classification. 

- In regression, target variable is a continuous variable.
    - Each leaf will correspond to a number. 

## How are Desision Trees Built ~4:08






Pure vs Impure - whether groups are divided homogeneously, that is, whether every member of that group is the same. 

impure - a group is not made up of all of one classification, but is mixed

pure - a group is made up of one classification, not mixed

worst kind of mixing is 50/50, best is 100/0

## Metrics on Impurity ~12:07

gini impurity - CART

entropy - c4.5

CHI-square statistical test - CHAID

for 50/50 - impurity is high

for 100/0 - impurity low

## CART - Example

branch on all three independent variables: gender, occupation, age. take gini impurity of the root node, and gini impurity of each potential branch (gender, occupation, age). select that variable as the starting branch that yields the best gini impurity improvement from the root node.

cross-validation - split data into 10 for example, then do a test/train split, alternating splits, average out errors, finally select tree with lowest cross-validation error value.
