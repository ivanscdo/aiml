```python
import pandas as pd
```

```python
data1 = pd.read_csv('files/Melbourne_housing_FULL.csv')
data1 = data1.dropna() # remove null values
data1.head()
```

## One Hot Encoding

```python
df_dummies = pd.get_dummies(data1, prefix='Park', columns=['ParkingArea'])
# replace categorical variable with numerical variable based on whether that categorial value is present in the record or not
# one hot encoding - encoding of categorical variables into 0 or 1
# ParkingArea variable turns into all values that it could have taken, now prefixed by park
# qualitative variable turned into quantitative
```

```python
df_dummies.head()
```

```python
df_dummies.columns
```

## Label Encoding
helps understand which variable got which name


```python
# from sklearn.prepreprocesing import LabelEncoder # this import did not work
# labelencoder = LabelEncoder()

from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()

df_dummies['RegionId'] = labelencoder.fit_transform(df_dummies.Regionname)
# transforms Regionname into 8 different columns as there were 8 values under the Regionname column
```

```python
df_dummies['RegionId'].head(5)
```

```python
df_dummies['RegionId'].unique()
```

```python
df_dummies['RegionId'].nunique()
```

```python
hotencoder = preprocessing.OneHotEncoder()
encoded = hotencoder.fit_transform(df_dummies.RegionId.values.reshape(-1,1)).toarray()
encoded
```

```python
encoded.shape
```

```python
# add the above back into dataframe
df_encoded = pd.DataFrame(encoded, columns = ['RegionId_'+str(int(i)) for i in range(encoded.shape[1])])
```

```python
df_encoded.head()
```

```python
df_encoded.shape
```

```python
df_dummies = pd.concat([df_dummies, df_encoded], axis=1) # concat two dataframes
```
