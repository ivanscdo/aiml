# Linear Regresssion 1

- linear regression - identify linear relationships between two variables

## measures of association
exercise:
- find center of scatterplot by finding x mean (xbar), y mean(ybar), and graphing the intersection
- divide scatterplot into quadrants using the point (xbar, ybar) as the origin
    1. points in quadrants 1 & 3 contribute to positive association
    1. points in quadrants 2 & 4 contribute to negative association
    
    
### 1. covariance
1. take a point(xi, yi), and the origin of the scatterplot (xbar, ybar) or the average/mean/center of all x and all y points.
1. take the product of the difference between (xi-xbar)(yi-xbar)
1. get the summed average
1. result can be from -infinity to +infinity
    - downward line - the smaller the number the greater the negative relationship
    - upward line - the larger the number, the greater the positive relationship


Similar to Variance formula
- note: Cov(x, x) = var(x)



major disadvantage: cannot independently tell if covariance is large is small since it is not unit-less

### 2. correlation (Pearson's correlation)

unit-less, units of x and units of y in numerator cancel with units in denominator



    - max: 1
    - min: -1
    - no correlation: 0
    - because it is unit free, it allows meaningfull comparisons
    - measure of linear relationship

## interpreting correlation

"*correlation does not imply causation*"

1. correlation - variables simply related to one another.
1. causation -  not only are they related, if one is changed, the other will change too. 

- x leading to y might be true, but cannot conclude that from data.
- data does not tell use causation. 

# Linear Regresssion 2

- simple model = equation
    - tells us how something works
    - give x (weight), get y (mpg)
    
    
- **linear regresssion - fitting a straight line through a scatterplot**


- difficult to know which line is the best fit
    - there can be an infinite number of lines that fit through the scatterplot, but which is the best?


- how to measure how good the line is?:
    - take point on line (predicted/y-i-hat), subtract by observed point (observed/y-i)
        - moving line can reduce size between a particular y-i-hat and y-i, but may increase size between other points
    - **residual = y-i(observed) - y-i-hat(predicted) = ei**
        - y-hat = predicted value
        - residuals of each possible line are summed and squared: **E ei^2**
            - squared because it removes signs
            - could use absolute value but it is incredibly difficult to handle mathematically
                - differential of absolute value is undefined
        - least squares of regression line
        - line with smallest E ei^2 is best line
        - since we are looking for a line (y=a+bx), it finds best a and best b. 

*19:03*
- knowing the best line does not tell us how the data looks
    - it does not tell us how close/far the data is to the line; how good is the regression fit?
    - best way is to look at how close/far the residuals are from the line
        - we do this by finding standard deviation of residuals
            - **tells us how the residuals are distributed around the line**
            - if std dev of residuals is small, good fit; if large, bad fit. 

*22:46*
- **standard deviation of residuals = standard error of regression (SE) = root mean square error (RMSE)**
    - how spread out are residuals around a straight line
- std dev of y before linear regression vs std dev of y after linear regression
    - before linear regression: E (yi - ybar)^2 **<- how spread out is y**
        - ybar = avg of y
    - after linear regression: E ei^2 **<- how spread out is y from the line**
    - both measure how much we were unable to explain. 
        - if they are equal, we didn't explain anything 
        - if E ei^2 is 0, we've explained everything

*27:25* - R^2
- what fraction/percentage of variation in y explained by the regression?
- R^2 = 1 - (E ei^2/E (y1-ybar)^2
    - 1 minus sum of ei squared divided by sum of yi minus ybar squared
    - perfect explanation: 1
    - explained nothing: 0
    - in-sample measure used in training data
    - out-of-sample R^2 - testing data
    

- sum of squared errors (SSE) = E ei^2
- sum of squared total (SST) = E (y1-ybar)^2

# Categorical Independent Variables in Regression

- categorical variables - strings, not ints

- regression will not work with categorical variables, must be encoded into number. 
- one-hot encoding - process of expanding column with categorical variables into multiple columns with either 1 or 0
- ex: Deptarment column has 3 values: Engineering, HR, Marketing. 
    - single column will be expanded into 3 columns Engineering, HR, and Marketing
    - 1 if row was previously marked as 1 of 3 values
    - 0 if row was not marked as 1 of 3 values
    - if row 1 was Engineering, it will have 1 for Engineering column, 0 for rest. 
- will choose one columns as base and leave it out



# Root Mean Square Error and Mean Absolute Error

### mean absolute error (MAE)
- 1/n (E |yi - yhat|)
    - averaged sum of the absolute value of the difference between observed value and predicted value
    
###  root mean square error (RMSE)
- ( 1/n (E (yi - yhat)^2 ) )**0.5
    - squared root sum of the square of the difference between observed value and predicted value
    
### MAE vs RMSE
- absolute values are difficult to differentiate
- squared root values are simple to differentiate 
    - easier to differentiate, easier to maximize/minimize
    

### Performance Measures on Training vs Testing Data
- every performance measure can be reduced by using more and more complex model
    - the more complex a model, the better fit it will have *on training data*
- all performance metrics will improve as model becomes more complicated
- **performance models should be tried on testing data**
    - if these performance measures used in training data, it will favor more complex models
