# 4. Importance of Scaling

## Issue with Eucladian Distance: Scaling
Eucladian - most important distance mneasure.
    
#### scale sensitivity:
- different columns with different units.
- ex: change from dollars to rupees, distance will change
- columns difference will dominate

## How to overcome issue?

computer __z-score__ - number of standard deviations away from the mean
- take xi value (old), substract out avg (M), divide by standard deviation



once scaled, all columns represent the same scale - standard deviations away from the mean

clustering critical dependence is on meausre of distance. 

once measure switched, even if scaled, will get different clusters!

okay to play with:
- distances
- iterations of clustering
- different k's
- number of clusters
- etc


playing around will benefit with better understanding of data

must understand data well!
- will implicitly contribute to better result of ML output 

## Disadvantages of Eucladian Distances

1. highly scaling dependent
    - can overcome by scaling
2. outliers
    - will skew mean, but also standard deviation
    - even when scaled, sensitivity to ouliers is never removed
    - Manhattan Distance is less sensitive to outliers

## Measures that Account for Relationship between Variables
1. Mahalanobis Distance
1. Jaccard Distance

### Mahalanobis Distance




Both points A and B are equidistant from center point, but only B is with the rest of the data, A is not. 

Need a way to tell how close A and B are from blue points (data). 

need to get a sense of how data is distributed, how are x and y related to one another.

Eucladian don't care about relation, simply take delta x and delta y (difference in x and difference in y)



Mahalonbis does the following: 
1. takes __principal components__ (eigen vectors of covariance variance Matrix)
1. natural axis of the data or how spread out the data are
1. draw ellipsis (contours) made up of equidistant points



### Jaccard Distance

1. measures distances between sets
1. \# of points at intersection of A & B, divided by total \# of poitns

# 6. Advantages & disadvantages of K- means Clustering

## Advantages

1. simple
    - no complex concepts; simply says: "if a bunch of points closer to each other, define points with respect to center (centroid)."
1. once centroids identified, assign new point to centroid
1. k-means technique, allows for the removal of subjectivity

## Disadvantages

1. how to choose k?
    - subjective
    - can play around, elbow method?
    - k can change as more data points flow in
1. k-mneans sensitive to initial centroid
    - start centroid, and iterate
    - depending on choice of first centroid, could have difference outcomes
    - non-convex optimization
1. curse of dimensionality
    - k-means depends on measure of distance
    - measure of distance depends on dimensionality
    - __as number of dimensions in which data points live becomes higher and higher, it becomes more and more computationality complex__

# 7. Silhouette coefficient for K means

#### Measure for how good the clustering is.

Silhouette Coefficient
- for each cluster, takes each point
- how close is each point to other points in same cluster
- compares to how close the point is to other clusters
- uses as measure of identifying how good the cluster is

## a(i) - similarities _within/inside_ a point's own cluster



##  b(i) - similarities _outside_ of a point's cluster



## How to calculate Silhouette Coefficient?

we will either have a coefficient:
- closer to 1 (ideal)
- 0 (not ideal)
- closer to -1 (not ideal)

# 9. Dynamic Clustering

In real world, __clustering does not happen on static data alone__.

Continuously done as data points come in. 

Advantages of clustering them over and over again?

might initially assign one customer to a particular cluster, but customer may move to one cluster to another. 

can keep track of how one customer moves from one segment to another?

Significant for merchant to understand what made the customer move. 

__However, first, we must identify if they moved.__

Dynamic Clustering  - clustering over and over again, while tracking:
1. how to data points move between clusters? 
1. why do they move?

Identify why some moved will help identify how to move other people. 

How do clusters themselves move? Why and how clusters moved.
